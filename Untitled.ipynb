# chatGPT API 사용해보기

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
```

```python
# 환경변수 로드
load_dotenv() # 현재 작업 디렉토리부터 .env 파일을 찾고 로드
```

```python
# OpenAI 모델 초기화
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.7, # temperature : 모델의 창의성을 조절하는 하이퍼파라미터
    api_key=os.getenv("OPENAI_API_KEY")
)
```

```python
# 테스트
response = llm.invoke('너는 뛰어난 화행분석가야. 다음의 문장에 대해 화행 분석을 수행해줘. “나는 서울 사는 26살이고 현재 미취업 상태야. 국취제 1 유형을 듣고 있고 교육 수당은 총 100만원을 받고 있는데 내가 받을 수 있는 정책좀”')
```

```python
print(f'응답 타입: {type(response)}')
print(f'응답 내용: {response}')
print(f'응답 텍스트만: {response.content}')
```

```python
print("\n=== 사용량 정보 ===")
if hasattr(response, 'usage_metadata'):
    for key, value in response.usage_metadata.items():
        print(f"  {key}: {value}")
```

## URL Loader
langchain은 웹에 기록된 글도 텍스트 형식으로 가져와 LLM에 활용할 수 있습니다.
대표적인 URL Loader는 WebBaseLoader와 UnstructuredURLLoader가 있습니다.

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader('https://news.nate.com/view/20250403n13734', encoding='cp949')

data = loader.load()

# 로더는 list 형태로 반환합니다. 
# 따라서, 인덱스를 사용하여 element에 접근하여 사용합니다.
print(f'로더의 반환 타입은? {type(data)}, {len(data)}') # list 타입
print(f'로더의 요소 타입은? {type(data[0])}') # langchain 패키지의 document 타입
print(data[0].page_content) 

# URL 로더는 웹 페이지의 모든 텍스트를 긁어온다는 특징이 있습니다.
# => 원치 않는 데이터까지 포함될 수 있음.

# bs4 없는 경우 pip install bs4
# 크롤링을 위한 라이브러리입니다.
```

```python
# 두 번째는 UnstrucuredURLLoader 입니다.
from langchain.document_loaders import UnstructuredURLLoader

urls = [
    'https://www.aitimes.com/news/articleView.html?idxno=200573',
    'https://news.nate.com/view/20250403n13734'
]

loader = UnstructuredURLLoader(urls)
data = loader.load() 
# 마찬가지로 모든 텍스트를 긁어온다는 특징이 있습니다.
```

# Text Splitter
키워드: 청크, 청킹, 토큰, 분할
토큰 제한이 있는 LLM이 여러 문장 덩어리(청크)를 참고해 답변할 수 있도록, 문장을 분할하는 역할.
이 청크 하나가 하나의 벡터가 된다.

Recursive < 
문맥유지를 위한 Chunk Overlap

LLM에는 토큰 제한이 있다. (최신 모델들은 대부분 16K지만..)
적절한 크기로 청킹해야하는 이유 : 각 청크는 사용자 질문에 더해져 일종의 context로 추가된다. 이 과정에서 청크의 크기가 너무 큰 경우, 토큰 제한을 넘어가는 문제가 발생할 가능성이 있다. 추가로, 토큰량이 과금과 연결되는 만큼, 적절한 수준으로 되는 것이 중요할 것이다.

몇 개의 청크를 참고를 할 지, 그 청크의 사이즈는 어떻게 될지, , 그 청크를 나눌 때 구분은 어떤 식으로 할지, 등..

대부분의 경우
RecursiveCharacterTextSplitter 사용 << 줄바꿈, 마침표, 쉼표 순으로 재귀적으로 분할 => max token 잘 지킴.
CharacterTextSplitter << 얘는 지정된 구분자 한 개만을 사용하므로, max_token을 지키지 못하는 경우 발생,
